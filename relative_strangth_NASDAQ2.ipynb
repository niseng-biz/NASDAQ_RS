IBDのRSの計算方法

対象銘柄のパフォーマンス　＝　
((((C - C63) / C63)) * 0.4) + (((C - C126) / C126)) * 0.2) + (((C - C189) / C189)) * 0.2) + (((C - C252) / C252)) * 0.2)) * 100

C＝直近終値

C63＝63日前（3ヶ月）の終値 　13週

C126＝126日前（6ヶ月）の終値　26週

C189＝189日前（9ヶ月）の終値　 39週

C252＝252日前（12ヶ月）の終値  52週

これは、過去63日（3ヶ月）、126日（6ヶ月）、189日（9ヶ月）、252日（12ヶ月）の変化を平均化したものです。63日(3ヶ月)の変化は、他の3つの期間に比べ、平均値への影響力が2倍になっています。

上記の計算式に基づき、①対象銘柄と②比較対象銘柄（ここではNASDAQ = "QQQ"）の銘柄パフォーマンスを計算します。

RSは、①と②の変化率として計算できます。具体的な計算式は以下の通りです。

Relative Strength ＝（①－②）÷ ② × 100 

ここでは、株価が指数より強ければ>0、指数より弱ければ<0というRelative Strengthの値が得られます。

```python
def calc_RS(dataframe):
    return (dataframe - dataframe.shift(13)) / dataframe.shift(13) * 0.4 + (dataframe - dataframe.shift(26)) / dataframe.shift(26) * 0.2 +  (dataframe - dataframe.shift(39)) / dataframe.shift(39) * 0.2 +  (dataframe - dataframe.shift(52)) / dataframe.shift(52) * 0.2 
    
    #return (dataframe - dataframe.shift(4)) / dataframe.shift(4) * 0.2 + (dataframe - dataframe.shift(13)) / dataframe.shift(13) * 0.2 + (dataframe - dataframe.shift(26)) / dataframe.shift(26) * 0.2 +  (dataframe - dataframe.shift(39)) / dataframe.shift(39) * 0.2 +  (dataframe - dataframe.shift(52)) / dataframe.shift(52) * 0.2 
    
```

```python
# df_w_RS = (df_w - df_w.shift(13)) / df_w.shift(13) * 0.4 + (df_w - df_w.shift(26)) / df_w.shift(26) * 0.2 +  (df_w - df_w.shift(39)) / df_w.shift(39) * 0.2 +  (df_w - df_w.shift(52)) / df_w.shift(52) * 0.2 
# df_w_RS = (df_w - df_w.shift(4)) / df_w.shift(4) * 0.2 + (df_w - df_w.shift(13)) / df_w.shift(13) * 0.2 + (df_w - df_w.shift(26)) / df_w.shift(26) * 0.2 +  (df_w - df_w.shift(39)) / df_w.shift(39) * 0.2 +  (df_w - df_w.shift(52)) / df_w.shift(52) * 0.2 
df_NASDAQ_w_RS = calc_RS(df_NASDAQ_w)
df_SP500_w_RS = calc_RS(df_SP500_w)
df_TSE_w_RS = calc_RS(df_TSE_w)
```

```python
df_NASDAQ_w_RS = df_NASDAQ_w_RS.div(df_NASDAQ_w_RS[ref], axis=0) *100
df_SP500_w_RS = df_SP500_w_RS.div(df_SP500_w_RS[ref], axis=0) *100
df_TSE_w_RS = df_TSE_w_RS.div(df_TSE_w_RS[ref_tse], axis=0) *100
```

```python
df_NASDAQ_w_RS.dropna(subset = ref, inplace = True)
df_SP500_w_RS.dropna(subset = ref, inplace = True)
df_TSE_w_RS.dropna(subset = ref_tse, inplace = True)
df_NASDAQ_w_RS
df_TSE_w_RS
```

```python
top20_list_NASDAQ = df_NASDAQ_w_RS.iloc[-1,:].sort_values(ascending=False).head(50)
top20_list_NASDAQ.rename("Relative Strength", inplace=True)
```

```python
top20_list_SP500 = df_SP500_w_RS.iloc[-1,:].sort_values(ascending=False).head(50)
top20_list_SP500.rename("Relative Strength", inplace=True)
```

```python
top20_list_TSE = df_TSE_w_RS.iloc[-1,:].sort_values(ascending=False).head(20)
top20_list_TSE.rename("Relative Strength", inplace=True)
```

```python
top20_df_NASDAQ = pd.DataFrame({'Ticker':top20_list_NASDAQ.index, 'Relative Strength':top20_list_NASDAQ.values})
top20_df_SP500 = pd.DataFrame({'Ticker':top20_list_SP500.index, 'Relative Strength':top20_list_SP500.values})
top20_df_TSE = pd.DataFrame({'Ticker':top20_list_TSE.index, 'Relative Strength':top20_list_TSE.values})
```

```python
top20_df_NASDAQ_state = pd.merge(top20_df_NASDAQ, ticker_df_NASDAQ, left_on='Ticker', right_on='Ticker')
top20_df_NASDAQ_state
```

```python
top20_df_SP500_state = pd.merge(top20_df_SP500, ticker_df_SP500, left_on='Ticker', right_on='Symbol').drop(columns = ['Symbol','Headquarters Location','Date added','CIK','Founded'] ).rename({'Security': 'Company'}, axis='columns')
top20_df_SP500_state
```

```python
top20_df_TSE_state = pd.merge(top20_df_TSE, ticker_df_TSE, left_on='Ticker', right_on='コード').drop(columns='コード')
top20_df_TSE_state
```

```python
merged_df =  pd.concat([top20_df_NASDAQ_state, top20_df_SP500_state]).sort_values('Relative Strength', ascending=False)
merged_df2 = merged_df.drop_duplicates(subset='Ticker').reset_index(drop=True).head(50)
merged_df2
```
